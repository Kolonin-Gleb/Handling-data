```python
from os import listdir
from shutil import copyfile

from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = '/content/drive/MyDrive/Colab\ Notebooks/Machine_learning/Datasets/cars.zip' # Путь к архиву датасета
dataset_dir = '/content/cars' # Место для разархивации
out_dir = '/content/all_cars' # Итоговое место. Все машины в одной папке
```

```python
!unzip -q {dataset} -d {dataset_dir}
```

```python
# Удаление директории с распакованными данными (если она есть)
!rm -rf {out_dir}
# Создание директории с распакованными данными
!mkdir {out_dir}
```

```python
labels = listdir(dataset_dir)
print(labels)
```

```python
for i in range(len(labels)):
    label = labels[i] # Текущий лейбл
    print(i, label)
    files = listdir(dataset_dir + "/" + label)
    for i_file in range(len(files)):
        f = files[i_file] # получение имени файла
        ext = f.split('.')[-1] # Взятие расширения файла
        fname = f"{i}_{i_file}.{ext}" # Новое имя файла

        # Использую функцию для переименования файла через копирование
        copyfile(dataset_dir+"/"+label+"/"+f, out_dir+"/"+fname)
```
